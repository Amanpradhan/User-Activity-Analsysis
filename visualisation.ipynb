```python
# Since we have more than 50% of device_type values as null we cannot remove it
```

```python
csv1['freq'] = csv1.groupby('device_type')['device_type'].transform('count')
```

```python
csv1
```

```python
# Alternate approach: we can either fill it with mean or median. 
# 1. Median
```

```python
csv1['device_type'] = csv1['device_type'].fillna(csv1['device_type'].median())
```

```python
csv1
```

```python
csv1['freq'] = csv1.groupby('device_type')['device_type'].transform('count')
```

```python
csv1
```

```python
# Now that we've replaced it with the median, our data is preprocessed.
```
